#### Line plots
~
Use the .iplot() method to generate a line plot with the dataset. This plot allows us to click on the elements in the legend to hide and display context which is pretty neat.There is a cursor to the top right of the plot to observe the various features of the plot. We can also use the zoom feature of specific areas of the plot.

```python
df.iplot()
```

#### Scatter plots
~
Use the .iplot() method with arguments kind (plot type), (x-axis variable), (y-axis variable), and mode argument removes the line connections setup by default with plotly. The plot can be zoomed in or out depending on need.

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers')
```

#### Bar plots
~
Use the .iplot() method with arguments kind (defines plot type), and an aggregate method to group data by. When we hover over the content we are able to view the actual data.

```python
df.sum().iplot(kind='bar')
```

### Boxplot
Use the .iplot() method with argument kind (defines plot type). Turn on and off data by selecting the specific element.

```python
df.iplot(kind='box')
```

### Surface plots
~
Use the .iplot() method with arguments kind (defines plot type), and colorscale to alter the plot color.

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

#### Histogram plots
~
 Use the .iplot() method with arguments kind (define plot type), and bins to specify the quantity of distribution buckets.

```python
df['A'].iplot(kind='hist',bins=50)
```

### Spread plot

```python
df[['A','B']].iplot(kind='spread')
```

### Bubble Plot

```python
df.iplot(kind='bubble',x='A',y='B',size='C')
```

#### Scatter matrix plot

```python
df.scatter_matrix()
```
